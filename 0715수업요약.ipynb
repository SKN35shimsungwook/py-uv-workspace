## 2. 반복되는 실수 패턴 (가장 중요)

오늘 하루 발생한 에러의 **대부분이 아래 4가지 패턴 중 하나**였습니다. 개념을 몰라서가 아니라, 아래 습관들 때문에 시간을 많이 쓰셨어요.

### 패턴 ① 오타 (전체 에러의 절반 이상)
| 의도한 것 | 실제로 쓴 것 |
|---|---|
| `__init__` | `__int__` |
| `getArea` | `gatArea` |
| `getTitle` | `gerTitle` |
| `toString` (클래스 안에 위치) | 클래스 밖으로 이탈 |
| `fread.close()` | `ferad.close()` |
| `if not line:` | `lf not line:` |
| `json_default_handler` | `json_defailt_handler` / `json_default_hadler` (정의·호출 이름 불일치) |
| `Exception` | `Exception0` |
| `students` | `student` (단수/복수 혼동) |
| `print` | `ptrint` |

**→ 처방**: 함수/메서드 이름은 **직접 타이핑하지 말고 복사-붙여넣기**를 습관화하세요. 특히 정의부와 호출부는 반드시 이름이 100% 일치해야 합니다.

### 패턴 ② 들여쓰기 오류
- `with` 문 아래 들여쓰기 누락 → `IndentationError`
- `for` 문 아래 들여쓰기 누락 → `IndentationError`
- 메서드가 클래스 밖으로 이탈 (`toString`, `getArea` 사례) → `AttributeError`

**→ 처방**: `:`(콜론)으로 끝나는 줄(`if`, `for`, `while`, `def`, `class`, `with`, `try` 등) 다음 줄은 **예외 없이 들여쓰기가 필수**라는 규칙을 다시 한번 명확히 하고 넘어가시면 좋겠습니다.

### 패턴 ③ 딕셔너리/리스트 문법 혼동
- 딕셔너리 키:값에 `=` 사용 (`headr = [...]` → `headr: [...]`)
- 딕셔너리 키에 따옴표 누락 (`headr:` → `'headr':`)
- 리스트/딕셔너리 항목 사이 콤마 누락

**→ 처방**: `{}`(딕셔너리)는 `키: 값`, 변수 대입은 `변수 = 값` — 이 둘이 헷갈리고 있습니다. "중괄호 안에서는 콜론, 중괄호 밖에서는 등호"로 암기하시길 권장합니다.

### 패턴 ④ 파일 열기 모드 불일치
- pickle에 `'w'` 모드 사용 → `wb`로 수정 필요했던 사례
- JSON 파일 읽을 때 `encoding='utf-8'` 누락 → `UnicodeDecodeError`
- 파일이 비어있는 상태에서 읽기 시도 → `JSONDecodeError`, `['']` 결과

**→ 처방**: 파일을 열 때마다 스스로에게 "① 텍스트냐 바이너리냐(`w`/`wb`), ② 한글이 있으면 encoding을 넣었나, ③ 쓰기 코드를 먼저 실행했나"를 체크하는 습관이 필요합니다.

## 3. 개념은 이해했지만 "직접 만들기"에서 막힌 지점

가장 솔직하게 짚어야 할 부분입니다. `headr`/`students`를 딕셔너리 리스트로 재구성하는 문제에서, **다 풀어서 설명하면 이해되지만 스스로 처음부터 설계하는 것은 어려워하는 모습**이 뚜렷했습니다. 이건 문법을 몰라서가 아니라, **"문제 → 코드 설계"로 넘어가는 사고 훈련이 아직 부족한 것**입니다.

### 권장 학습 방법 (오늘 안내드린 3단계)
1. 코드 짜기 전에 **한국어로 순서를 먼저 적기**
2. 지금까지 배운 뼈대(패턴)에 지금 문제가 해당하는지 판단하기
3. 반복문 없이 **1개짜리로 먼저 성공**시킨 후 `for`로 확장하기

## 4. 개념별 이해도 체크리스트

| 개념 | 상태 | 비고 |
|---|---|---|
| 반복문, list comprehension | ✅ 양호 | 스스로 응용까지 시도함 (fruits 문제 등) |
| 클래스 변수 vs 인스턴스 변수 | ⚠️ 재확인 필요 | 처음엔 헷갈렸으나 비유(칠판/가방)로 이해 완료. 실전 적용에서 다시 헷갈릴 수 있음 |
| `__init__`, 메서드 정의 | ✅ 양호 | 오타만 조심하면 문제없음 |
| 예외처리 (`try-except-else-finally`) | ✅ 양호 | 다중 except 순서(자식→부모)까지 정확히 이해 |
| 사용자 정의 예외, `abstractmethod` | ✅ 양호 | 개념 이해는 잘 됨 |
| 파일 입출력 기본 | ✅ 양호 | 모드(`r`/`w`/`a`) 구분 잘함 |
| pickle vs json 구분 | ✅ 양호 | 보안/호환성 차이까지 이해 |
| **딕셔너리·리스트를 조합해서 새 구조 설계하기** | ❌ **가장 취약** | 이번 보고서의 핵심 보완 지점 |
| `zip()`, `dict()` 조합 응용 | ⚠️ 재확인 필요 | 결과 코드는 이해하지만 처음 설계는 어려움 |
| f-string / `.format()` | ✅ 양호 | 뒤늦게 질문했지만 바로 이해 |
