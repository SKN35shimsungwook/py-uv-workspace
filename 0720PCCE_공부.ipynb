```python
# 1단계: 출생 연도(year)와 나이 타입(age_type)을 입력받습니다.
year = int(input())
age_type = input()

# 2단계: age_type에 따라 한국식 나이 또는 연 나이를 계산합니다.
if age_type == "korea":
    # 한국식 나이: (기준 연도 - 출생 연도) + 1
    answer = 2030 - year + 1

elif age_type == "year":
    # 연 나이: (기준 연도 - 출생 연도)
    answer = 2030 - year

# 3단계: 최종 계산된 나이를 출력합니다.
print(answer)
```

```python
scores = [70, 85, 90, 55, 80, 65]
count = 0

# scores 리스트에 있는 점수를 하나씩 score 변수에 꺼내어 확인합니다.
for score in scores:
    # 점수가 80점 이상인지 확인합니다.
    if score >= 80:
        count = count + 1

print(count)
```

```python
scores = [70, 85, 90, 55, 80, 65]

# 첫 번째 점수로 max_score를 초기화합니다.
max_score = scores[0]

for score in scores:
    # 현재 점수(score)가 지금까지의 최고점(max_score)보다 큰지 확인합니다.
    if score > max_score:
        # 더 큰 점수를 발견하면 max_score를 해당 점수로 업데이트합니다.
        max_score = score

print(max_score)
```

```python
# 두 수 중 더 큰 값을 반환하는 함수 정의
def get_max(a, b):
    if a > b:
        return a
    else:
        return b

# 함수 호출 및 결과 출력
result = get_max(15, 23)
print(result)  # 23이 출력되어야 합니다.
```

```python
def get_average(numbers):
    # sum() 함수로 합계를 구하고, len() 함수로 개수를 구합니다.
    total = sum(numbers)
    length = len(numbers)
    
    # 평균 = 총합 / 개수
    avg = total / length
    return int(avg)

# 함수 테스트
sample_list = [10, 20, 30, 40]
print(get_average(sample_list))  # 25.0 이 출력되어야 합니다.
```

```python
def check_even_odd(n):
    # 2로 나눈 나머지가 0인지 확인합니다.
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

# 함수 테스트
print(check_even_odd(7))   # "Odd"가 출력되어야 합니다.
print(check_even_odd(10))  # "Even"이 출력되어야 합니다.
```
