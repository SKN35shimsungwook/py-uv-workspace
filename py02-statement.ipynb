# 식별 연산자 (Identity Operators)
- 식별 연산자는 두 객체가 같은 객체인지를 확인한다.
- is : 두 객체가 같은 객체인지 확인
- is not 두 객체가 다른 객체인지 확인
- 결과 : True, False


````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)  # True (a와 b는 같은 객체)
print(a is c)  # False (a와 c는 같은 값을 가지지만 다른 객체)
````


## 반복문 : while
반복문에서 else는 반복문이 완벽하게 끝났을 때 실행된다.  
대부분의 프로그래밍 언어는 while과 for을 지원한다.  
else
  - if ~ else    
  - while ~ else  
  - try ~ except ~ else ~ finally


```
while 조건식:
  수행할 문장
else:
  수행할 문장
```


````python
# while문이 완벽이  돌았으므로 end가 출력된다.
a = 10
while  a > 5 :
  a -= 1  # a = a - 1
  print(a)
else:
  print('end')
````

````python
# while문을 강제적으로 빠져나왔기 때문에 else영역을 실행하지 않는다.
a = 10
while  a > 5 :
  a -= 1  # a = a - 1
  print(a)
  if a==5:
    break   # 반복문을 강제적으로 빠져나올때 사용한다. 즉 반복문을 중단시킨다
else:
  print('end')
````

````python
#  continue는 다음 루프(조건식)로 바로 넘어간다.
# 반복문을 중단하지는 않으므로 else문이 실행된다.
a = 10 #5
while a > 5:
  a -= 1
  if a % 2 == 0 :
    continue   # 이동
  print(a)
else:
  print('end')
````

List Comprehension

   -A = [표현식 for 항목1 in 반복가능객체(iterable)1 for 항목2 in 반복가능객체(iterable)2 if 논리식]


````python
s= [] 
for x in range(10):
    s.append (x**2)
print(s)
````

````python
s = [x**2 for x in range(10)]
print(s)
````

 - 사전 정의시 값을 정하지 않고 호출시사전 내의 값들이 처리되도록 구성
 - A = {표현식 for (k, v) in 반복가능객체 if 논리식}



````python

d = {}
for k, v in zip(range(9), range(9)):
  d[k]=v
print(d)
````

````python
d = {k: v for k, v in zip(range(9), range(9))}

print(d)
````

### Generator Expression
 - 파이썬에는 tuple(immutable)에는 comprehension이 존재하지 않고
    syntax 상으로 generator 가 생성한다.
 - 값을 하나씩 “생성”하는 반복자(iterator). 메모리를 덜 사용하며, **지연 평가(Lazy Evaluation)**가 특징이다.
 - generator를 호출하면 next 메소드를 통해 하나씩 호출하며 StopIteration 이 발생하면 종료된다.


````python
squares = (x **2 for x in range(3))
print(squares)   # 이건 제너레이터 객체 생성 (실행 아님)
print(dir(squares))
````

````python
squares.__next__()
````

````python
print(next(squares))
````

````python
def count_up_t(n):
    i = 1
    while i <= n:
         yield i
         i += 1
         print(i)


data = count_up_t(3)
print(data)
````

만약 1,000만 개의 데이터 중 연산 비용이 아주 큰 작업(예: time.sleep이 포함되거나 복잡한 수학 연산)을 수행하다가, 원하는 조건(예: 3의 배수)을 가진 첫 번째 결과만 찾고 싶을 때 두 방식의 차이

````python
import time


# 무겁고 오래 걸리는 연산을 시뮬레이션하는 함수
def heavy_processing(n):
    # 실제로는 아주 복잡한 연산이 일어난다고 가정합니다.
    return n * 2


# 1. 조기 평가 (Eager Evaluation) 방식
def eager_approach(data):
    # 조건과 상관없이 1,000만 개 데이터 전체에 대해 heavy_processing을 먼저 '전부' 실행함
    result = [heavy_processing(x) for x in data]
   
    # 그 후, 필요한 조건을 찾음
    for val in result:
        if val > 0 and val % 3 == 0:
            return val


# 2. 지연 평가 (Lazy Evaluation) 방식
def lazy_approach(data):
    # 이 시점에는 아무런 연산도 하지 않음 (계산식만 준비)
    result_generator = (heavy_processing(x) for x in data)
   
    # 필요한 순간(next)에만 heavy_processing을 하나씩 실행함
    for val in result_generator:
        if val > 0 and val % 3 == 0:
            return val


# 테스트 데이터 (0부터 9,999,999까지)
data_list = range(10000000)


# [비교 시작]
# 1) 조기 평가는 1,000만 번의 heavy_processing을 모두 수행하느라 엄청난 시간이 걸림
# eager_approach(data_list)


# 2) 지연 평가는 단 4번(x=0,1,2,3)만 연산하고 x=3일 때 6(3의 배수)을 반환하며 즉시 종료!
start_time = time.time()
print(f"지연 평가 결과: {lazy_approach(data_list)}")
print(f"소요 시간: {time.time() - start_time:.6f}초")
````

````python
import time



def heavy_processing(n):
    return n * 2


def eager_approach(data):
    result = [heavy_processing(x) for x in data]
    for val in result:
        if val > 0 and val % 3 == 0:
            return val


def lazy_approach(data):
  
    result_generator = (heavy_processing(x) for x in data)
    for val in result_generator:
        if val > 0 and val % 3 == 0:
            return val



data_list = range(10000000)


start_time = time.time()
print(f"지연 평가 결과: {lazy_approach(data_list)}")
print(f"소요 시간: {time.time() - start_time:.6f}초")


start_time2  = time.time()
print(f"지연 평가 결과: {eager_approach(data_list)}")
print(f"소요 시간: {time.time() - start_time:.6f}초")
````

````python
def check_valid_input(user_input: str) -> bool:
    """
    사용자가 입력한 값이 '금지어 목록'에 걸리는지 확인하는 함수
    """
    # .strip().lower() -> 공백이나 대소문자 차이로 오작동하는 걸 방지
    # (예: "Admin ", "ADMIN", "admin" 모두 같은 취급을 받게 함)
    cleaned = user_input.strip().lower()

    forbidden_words = ["admin", "root", "system"]

    # any()로 forbidden_words 중 하나라도 cleaned 안에 포함돼있는지 체크
    return any(word in cleaned for word in forbidden_words)


# 테스트
try:
    test_value = "  Admin123  "
    if check_valid_input(test_value):
        print("⚠️ 사용 불가능한 아이디입니다.")
    else:
        print("✅ 사용 가능한 아이디입니다.")
except Exception as e:
    # 혹시 모를 예외(예: None이 들어왔을 때) 방어
    print(f"입력 처리 중 오류 발생: {e}")
````
