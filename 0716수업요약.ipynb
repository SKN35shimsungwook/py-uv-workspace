## 전체 구조 한눈에 보기 (마인드맵 형식)

```
Streamlit
│
├── 1. 개요
│   ├── 파이썬 코드만으로 웹앱 제작
│   ├── HTML/CSS/JS 불필요
│   └── 데이터 사이언스·ML 프로토타입에 최적화
│
├── 2. 기본 텍스트/제목 요소
│   ├── st.title()
│   ├── st.header()
│   ├── st.subheader()
│   └── st.write() / st.markdown()
│
├── 3. 입력 위젯 (사용자 입력받기)
│   ├── st.text_input()   → 문자열 입력
│   ├── st.number_input()  → 숫자 입력 (min/max 지정)
│   ├── st.selectbox()     → 드롭다운 선택
│   ├── st.checkbox()      → 체크박스
│   └── st.button()        → 클릭 이벤트
│
├── 4. 조건부 로직 & 상태
│   ├── if st.button(...): → 클릭 시 True 반환
│   ├── st.balloons()      → 이벤트 애니메이션
│   └── 버튼 옵션: type(primary/secondary), disabled, use_container_width
│
├── 5. 레이아웃 구성
│   ├── st.sidebar         → 사이드바 영역
│   ├── st.columns([비율]) → 화면을 여러 열로 분할
│   └── st.divider()       → 구분선
│
├── 6. 파일/이미지 다루기
│   ├── st.file_uploader() → 파일 업로드 받기
│   ├── open(path, 'rb')   → 이미지 파일 직접 열기
│   └── pd.read_csv()      → 업로드된 CSV 읽기 (encoding 주의)
│
├── 7. 데이터 시각화
│   ├── st.bar_chart()     → 막대 차트
│   ├── st.line_chart()    → 꺾은선 차트
│   ├── st.table()         → 표 형태 데이터
│   └── pandas DataFrame과 연동
│
├── 8. 고급 커스터마이징
│   ├── st.markdown(..., unsafe_allow_html=True) → 직접 HTML/CSS 삽입
│   ├── st.echo()          → 코드 블록 그대로 표시
│   └── st.spinner()       → 로딩 중 표시
│
└── 9. 실행 & 배포
    ├── streamlit run app.py
    └── %%writefile (주피터에서 .py 파일로 저장 후 실행)
```

## 요점 정리 (핵심 개념별)

### ① Streamlit의 정체성
- **파이썬 코드 → HTML 자동 변환** 프레임워크
- 프론트엔드 지식(HTML/CSS/JS) 없이 웹앱 제작 가능
- 내부적으로는 React 프론트엔드 + 파이썬 백엔드로 구성되어 있으나, 사용자는 이 구조를 몰라도 됨

### ② 기본 텍스트 요소
| 함수 | 용도 |
|---|---|
| `st.title()` | 페이지 제목 |
| `st.header()` / `st.subheader()` | 섹션 제목 |
| `st.write()` | 범용 출력 (텍스트, 데이터프레임 등 자동 인식) |
| `st.markdown()` | 마크다운 문법 지원, `unsafe_allow_html=True`로 HTML도 삽입 가능 |

### ③ 입력 위젯
- `st.text_input(라벨)` — 문자열 입력창
- `st.number_input(라벨, min_value=, max_value=)` — 범위 지정 숫자 입력
- `st.selectbox(라벨, 옵션리스트)` — 드롭다운
- `st.checkbox(라벨)` — 체크박스, `True`/`False` 반환

### ④ 버튼과 조건 처리
```python
if st.button("누르세요"):
    st.balloons()
    st.write("버튼이 눌렸습니다!")
```
- 버튼은 클릭 시 **그 순간만** `True`를 반환 (상태가 유지되지 않음 — Streamlit의 재실행 모델 특성)
- 옵션: `type="primary"/"secondary"`, `disabled=True/False`, `use_container_width=True/False`, `key=`(위젯 구분용 고유값)

### ⑤ 레이아웃
- **사이드바**: `st.sidebar` 또는 `with st.sidebar:` 블록으로 왼쪽 고정 영역 구성
- **컬럼 분할**: `col1, col2 = st.columns([3,1])` → 비율(3:1)로 화면을 나눔
- 각 컬럼 객체(`col1`, `col2`)에 `.subheader()`, `.markdown()`, `.bar_chart()` 등을 직접 호출

### ⑥ 데이터 업로드 & 처리
- `st.file_uploader()`로 파일을 받고, `pd.read_csv()`로 읽음
- **한국어 CSV는 `encoding='cp949'` 또는 `'euc-kr'`을 명시해야 하는 경우가 많음** (오늘 직접 겪은 에러)
- 이미지 파일은 `open(path, 'rb')`로 열거나 `st.image(path)`로 바로 표시

### ⑦ 시각화
- `st.bar_chart(data)`, `st.line_chart(data)` — pandas 데이터를 바로 넣어 시각화
- `st.table(df)` — DataFrame을 표 형태로 출력
- 사이드바의 `selectbox`로 "막대/꺾은선/히스토그램" 등 차트 종류를 선택하게 만드는 패턴 자주 사용

### ⑧ 고급: HTML/CSS 커스터마이징
```python
st.markdown('''
<style>
.custom-column { background-color: lightblue; padding: 5px; }
</style>
''', unsafe_allow_html=True)

col1.markdown('<div class="custom-column">', unsafe_allow_html=True)
```
- Streamlit 기본 디자인을 넘어서고 싶을 때, **파이썬 코드 안에서 CSS를 문자열로 주입**
- `unsafe_allow_html=True`를 켜야 HTML로 해석됨 (기본은 텍스트로만 표시)

### ⑨ 실행 방법
- 주피터에서 `%%writefile app.py`로 코드를 파일로 저장
- 터미널에서 `streamlit run app.py`로 실제 웹앱 실행 (브라우저 자동 오픈)

## 2. 개념별 정리 (쉬운 설명 포함)

### 2-1. Streamlit이란 무엇인가
파이썬 코드만 짜면, Streamlit이 그걸 자동으로 HTML 웹페이지로 바꿔주는 도구입니다. 비유하면 "통역사"예요 — 개발자는 파이썬으로만 말하고, Streamlit이 브라우저가 알아듣는 말(HTML/CSS/JS)로 실시간 통역해줍니다. 그래서 HTML/CSS/JS를 몰라도 웹앱을 만들 수 있습니다.

- 2018~2019년 출시, 이미 7년 된 검증된 도구
- 내부적으로는 React(프론트) + 파이썬(백엔드) 구조지만, 사용자는 이걸 몰라도 됨
- 데이터 앱/프로토타입/내부 도구에 최적화 — 대규모 상용 서비스는 여전히 전통 방식(프론트+백엔드 분리) 사용

### 2-2. 기본 출력 & 입력 위젯
| 함수 | 역할 |
|---|---|
| `st.title/header/subheader` | 제목 표시 |
| `st.write/markdown` | 텍스트 출력, `markdown`은 HTML도 삽입 가능(`unsafe_allow_html=True`) |
| `st.text_input/number_input/selectbox/checkbox/radio/slider` | 각각 문자열/숫자/드롭다운/체크박스/라디오/슬라이더 입력 |

**핵심 원리**: 위젯 값을 받는 변수(`name = st.text_input(...)`)는 사용자가 입력할 때마다 **앱 전체가 처음부터 다시 실행**되면서 갱신됩니다. 이게 Streamlit의 근본 동작 방식입니다.

### 2-3. 버튼과 조건 처리
```python
if st.button("누르세요"):
    st.balloons()
```
- 버튼은 눌리는 "그 순간만" `True`를 반환합니다. 클릭 후 다시 스크립트가 재실행되면 버튼 상태는 초기화됩니다.
- 옵션: `type`(primary/secondary), `disabled`, `use_container_width`, `key`(위젯 구분용 고유 이름표)
- `on_click=함수`로 클릭 시 실행할 함수를 미리 연결할 수도 있음 (`args=`, `kwargs=`로 인자 전달)

### 2-4. 레이아웃 구성
- **사이드바**: `st.sidebar.제목()` 또는 `with st.sidebar:` 블록
- **컬럼 분할**: `col1, col2 = st.columns([3,1])` → 비율대로 화면을 나눔, 이후 `col1.write(...)`처럼 각 컬럼 객체에 직접 호출
- **탭**: `tab1, tab2 = st.tabs(["A","B"])`, `with tab1:` 블록 안에 그 탭의 내용 작성

### 2-6. 데이터 시각화

### 2-7. 세션 상태 (`st.session_state`) — 오늘 새로 배운 핵심 개념
Streamlit은 위젯을 조작할 때마다 전체 코드를 처음부터 다시 실행합니다. 그러면 "지금까지 누른 횟수" 같은 정보가 매번 사라지겠죠. 이를 막기 위해 **재실행되어도 값이 유지되는 저장소**가 `st.session_state`입니다.

```python
if "count" not in st.session_state:
    st.session_state.count = 0
if st.button("증가"):
    st.session_state.count += 1
st.write(st.session_state.count)
```
비유: 스크립트가 매번 리셋되는 "일회용 메모지"라면, `session_state`는 "계속 펴놓고 쓰는 노트"입니다.

### 2-8. 폼 (`st.form`)
```python
with st.form('myform'):
    name = st.text_input('이름')
    submit = st.form_submit_button('확인')
    if submit:
        st.write(name)
```
- 폼 안의 위젯들은 **제출 버튼을 누르기 전까지는 재실행을 유발하지 않고** 값을 모아뒀다가, 제출 시 한꺼번에 처리합니다. 여러 입력을 받을 때 화면이 계속 깜빡이며 재실행되는 걸 방지하는 장치입니다.

### 2-11. 정규표현식 기초 (`re` 모듈)
```python
import re
ta = re.sub('[^a-zA-Z]', ' ', text)   # 영문자 아닌 것 → 공백
re.sub(r'[\s]+', ' ', ta)              # 연속 공백 → 공백 하나
```

## 4. 오늘 발생한 오류/혼동 정리 및 원인 분석

| 발생 상황 | 에러/증상 | 근본 원인 |
|---|---|---|
| `#` 주석 작성 | `NameError: name '기본값' is not defined` | 주석(`#`)과 설명 텍스트가 줄바꿈으로 분리되어, 다음 줄이 코드로 해석됨 |
| CSV 업로드 읽기 | `UnicodeDecodeError` | `pd.read_csv()` 기본 인코딩(UTF-8)과 실제 파일 인코딩(cp949) 불일치 |
| 이미지 경로 | `FileNotFoundError: 'images/cat.jpg'`, `'./souce/cat.jpg'` | 폴더 자체가 없거나, `source`를 `souce`로 오타 |
| 코드 마무리 실수 | `SyntaxError: invalid syntax` (문자열 뒤 `z`) | 타이핑 중 불필요한 글자가 끼어듦 |
| `with` 블록 구성 | `IndentationError: expected an indented block after 'with'` (반복) | `with col1:`, `with st.spinner(...):` 아래 실행 코드 누락 |
| HTML 삽입 | `SyntaxError`, `AttributeError` | `<'div calss=...`처럼 따옴표 위치 오류 + `class` 오타(`calss`) |
| CSS 작성 | (논리 오류, 에러는 아님) | `pading` → `padding` 오타 |
| 컬럼 변수 사용 | (논리 오류) | `col2` 영역인데 마지막 줄만 `col1.table(df)`로 복사 후 이름 미변경 |
| 학사정보검색 조건문 | `NameError`, `AttributeError`(makedown, deader) | `if st.markdown(...)`, `if st.header(...)`처럼 출력 함수를 조건문으로 오용 + 함수명 오타 |
| 검색 조건 확장 | `SyntaxError: expected ':'` (반복) | `or`/`and` 결합 시 괄호 안 닫힘, if문 콜론 누락 |
| 여러 문법 오류 중첩 | `IndentationError`, 콜론 중복 등 다발 | `\` 줄연결 문법 오용, `else`에 조건을 붙이려 시도(`else not found:`), `st.write(...):`처럼 함수 호출 뒤 불필요한 콜론 |
| 반복 수정 후 재실행 | `IndentationError: unindent does not match any outer indentation level` | 탭과 스페이스 혼용 추정 |
| Git 커밋 | 진행 안 되고 로딩만 지속 | `COMMIT_EDITMSG` 편집기 탭이 열린 채 저장/닫기를 안 함 |
| 정규표현식 사용 | `NameError: name 're' is not defined` | `import re` 누락 |

## 5. 취약점 분석

### 취약점 ① "출력 함수"와 "조건/판단 함수"를 구분하지 못함 (오늘 가장 두드러진 패턴)
`st.markdown()`, `st.header()`, `st.write()`는 **화면에 뭔가를 보여주기만** 하는 함수인데, 이를 `if` 조건으로 감싸는 시도가 여러 차례 반복되었습니다(`if st.markdown('홍길동'):`, `if st.header('홍길동'):`). 실제 조건 판단이 필요한 자리에는 **비교 연산(`==`), 논리 연산(`and/or`), `True/False`를 반환하는 함수(`st.button()` 등)**가 와야 한다는 구분이 아직 확실히 자리잡지 않았습니다.

### 취약점 ② 복합 조건문(`and`/`or`/괄호) 조립 시 문법 오류 다발
학사정보검색 앱의 검색 조건을 여러 항목(학번/이름/전공) 결합으로 확장하는 과정에서, 괄호 여닫음 불일치, 콜론 누락, `\` 줄연결 오용, `else`에 조건을 붙이려는 시도 등 **문법 요소가 한꺼번에 여러 개 틀리는 상황**이 발생했습니다. 하나씩 순서대로 쌓기보다, 여러 수정을 동시에 시도하다 꼬인 것으로 보입니다.

### 취약점 ③ 오타 (어제부터 이어지는 패턴)
`makedown`(→markdown), `deader`(→header), `pading`(→padding), `calss`(→class), `souce`(→source) 등, **철자 일부가 바뀌거나 빠지는 오타**가 오늘도 여러 차례 발생했습니다.

### 취약점 ④ 인코딩/경로 문제
한글 CSV의 인코딩 불일치, 파일 경로 오타로 인한 `FileNotFoundError`가 반복되었습니다. 파일을 다루는 코드를 짤 때 "인코딩을 명시했는가", "경로가 실제로 존재하는가"를 사전에 확인하는 습관이 아직 부족합니다.

### 취약점 ⑤ 새 개념(session_state, form) 자체는 코드 읽기는 가능하나, 왜 필요한지 체화는 진행 중
MBTI 코드를 리뷰하며 `session_state`, `st.form`의 동작을 설명 듣고 이해했지만, 직접 학사정보검색 앱을 만들 때는 이 개념들을 자발적으로 활용하지 않았습니다. 개념 이해와 실전 적용 사이에 아직 간격이 있습니다.

### 취약점 ⑥ Git 커밋 절차(GUI 도구 사용법)에 대한 낯섦
코드 문법과 무관하게, VS Code의 `COMMIT_EDITMSG` 편집기 흐름(입력→저장→탭 닫기)을 몰라 커밋이 멈춘 것처럼 느껴진 사례가 있었습니다. 이는 프로그래밍 개념이 아니라 **도구 사용법**의 문제입니다.

## 6. 복습 우선순위 제안

| 순위 | 항목 | 복습 방법 |
|---|---|---|
| 1 | 출력 함수 vs 조건 판단 구분 | `st.write/markdown/header`는 절대 `if`로 감싸지 않는다는 규칙을 먼저 확정하고, `st.button()`처럼 True/False를 반환하는 함수만 조건에 쓰는 연습 |
| 2 | 복합 조건문 조립 연습 | `(A and 조건1) or (B and 조건2)` 형태를 괄호 개수를 세어가며 한 번에 하나씩만 수정하는 습관 |
| 3 | 오타 예방 | 함수 이름은 Streamlit 공식 문서에서 복사하거나, 자동완성(VS Code IntelliSense) 적극 활용 |
| 4 | 파일 경로/인코딩 사전 점검 습관 | 파일 다루는 코드 작성 시 `encoding='utf-8'` 기본 명시 + 경로 존재 여부를 `os.path.exists()`로 먼저 확인 |
| 5 | `session_state`/`form` 직접 적용해보기 | 학사정보검색 앱에 검색 이력(session_state)을 추가하는 식으로 스스로 응용 연습 |
| 6 | Git 커밋 절차 숙달 | `COMMIT_EDITMSG` 흐름을 한 번 더 의식적으로 연습(터미널 `git commit -m "메시지"` 방식도 함께 익히면 편집기 이슈 회피 가능) |

## 7. 총평

오늘은 어제 다진 기초 문법을 실제 동작하는 웹앱(Streamlit)으로 확장한 하루였습니다. 학사정보검색 앱과 MBTI 퀴즈 코드를 통해 **위젯, 레이아웃, 세션 상태, 폼**이라는 Streamlit 핵심 개념을 다뤘고, 특히 학사정보검색 앱은 여러 차례의 시행착오 끝에 스스로 완성했습니다. 가장 눈에 띄는 개선 포인트는 **"화면에 보여주는 함수"와 "조건을 판단하는 로직"을 구분하는 감각**이며, 이는 앞으로 Streamlit 앱을 만들 때마다 반복해서 부딪힐 가능성이 높으니 이번 주 안에 다시 한번 짚고 넘어가는 것을 권장합니다. 나머지(오타, 인코딩, Git 절차)는 반복 사용을 통해 자연히 개선될 성격의 문제입니다.
