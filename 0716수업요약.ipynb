## 전체 구조 한눈에 보기 (마인드맵 형식)

```
Streamlit
│
├── 1. 개요
│   ├── 파이썬 코드만으로 웹앱 제작
│   ├── HTML/CSS/JS 불필요
│   └── 데이터 사이언스·ML 프로토타입에 최적화
│
├── 2. 기본 텍스트/제목 요소
│   ├── st.title()
│   ├── st.header()
│   ├── st.subheader()
│   └── st.write() / st.markdown()
│
├── 3. 입력 위젯 (사용자 입력받기)
│   ├── st.text_input()   → 문자열 입력
│   ├── st.number_input()  → 숫자 입력 (min/max 지정)
│   ├── st.selectbox()     → 드롭다운 선택
│   ├── st.checkbox()      → 체크박스
│   └── st.button()        → 클릭 이벤트
│
├── 4. 조건부 로직 & 상태
│   ├── if st.button(...): → 클릭 시 True 반환
│   ├── st.balloons()      → 이벤트 애니메이션
│   └── 버튼 옵션: type(primary/secondary), disabled, use_container_width
│
├── 5. 레이아웃 구성
│   ├── st.sidebar         → 사이드바 영역
│   ├── st.columns([비율]) → 화면을 여러 열로 분할
│   └── st.divider()       → 구분선
│
├── 6. 파일/이미지 다루기
│   ├── st.file_uploader() → 파일 업로드 받기
│   ├── open(path, 'rb')   → 이미지 파일 직접 열기
│   └── pd.read_csv()      → 업로드된 CSV 읽기 (encoding 주의)
│
├── 7. 데이터 시각화
│   ├── st.bar_chart()     → 막대 차트
│   ├── st.line_chart()    → 꺾은선 차트
│   ├── st.table()         → 표 형태 데이터
│   └── pandas DataFrame과 연동
│
├── 8. 고급 커스터마이징
│   ├── st.markdown(..., unsafe_allow_html=True) → 직접 HTML/CSS 삽입
│   ├── st.echo()          → 코드 블록 그대로 표시
│   └── st.spinner()       → 로딩 중 표시
│
└── 9. 실행 & 배포
    ├── streamlit run app.py
    └── %%writefile (주피터에서 .py 파일로 저장 후 실행)
```

## 요점 정리 (핵심 개념별)

### ① Streamlit의 정체성
- **파이썬 코드 → HTML 자동 변환** 프레임워크
- 프론트엔드 지식(HTML/CSS/JS) 없이 웹앱 제작 가능
- 내부적으로는 React 프론트엔드 + 파이썬 백엔드로 구성되어 있으나, 사용자는 이 구조를 몰라도 됨

### ② 기본 텍스트 요소
| 함수 | 용도 |
|---|---|
| `st.title()` | 페이지 제목 |
| `st.header()` / `st.subheader()` | 섹션 제목 |
| `st.write()` | 범용 출력 (텍스트, 데이터프레임 등 자동 인식) |
| `st.markdown()` | 마크다운 문법 지원, `unsafe_allow_html=True`로 HTML도 삽입 가능 |

### ③ 입력 위젯
- `st.text_input(라벨)` — 문자열 입력창
- `st.number_input(라벨, min_value=, max_value=)` — 범위 지정 숫자 입력
- `st.selectbox(라벨, 옵션리스트)` — 드롭다운
- `st.checkbox(라벨)` — 체크박스, `True`/`False` 반환

### ④ 버튼과 조건 처리
```python
if st.button("누르세요"):
    st.balloons()
    st.write("버튼이 눌렸습니다!")
```
- 버튼은 클릭 시 **그 순간만** `True`를 반환 (상태가 유지되지 않음 — Streamlit의 재실행 모델 특성)
- 옵션: `type="primary"/"secondary"`, `disabled=True/False`, `use_container_width=True/False`, `key=`(위젯 구분용 고유값)

### ⑤ 레이아웃
- **사이드바**: `st.sidebar` 또는 `with st.sidebar:` 블록으로 왼쪽 고정 영역 구성
- **컬럼 분할**: `col1, col2 = st.columns([3,1])` → 비율(3:1)로 화면을 나눔
- 각 컬럼 객체(`col1`, `col2`)에 `.subheader()`, `.markdown()`, `.bar_chart()` 등을 직접 호출

### ⑥ 데이터 업로드 & 처리
- `st.file_uploader()`로 파일을 받고, `pd.read_csv()`로 읽음
- **한국어 CSV는 `encoding='cp949'` 또는 `'euc-kr'`을 명시해야 하는 경우가 많음** (오늘 직접 겪은 에러)
- 이미지 파일은 `open(path, 'rb')`로 열거나 `st.image(path)`로 바로 표시

### ⑦ 시각화
- `st.bar_chart(data)`, `st.line_chart(data)` — pandas 데이터를 바로 넣어 시각화
- `st.table(df)` — DataFrame을 표 형태로 출력
- 사이드바의 `selectbox`로 "막대/꺾은선/히스토그램" 등 차트 종류를 선택하게 만드는 패턴 자주 사용

### ⑧ 고급: HTML/CSS 커스터마이징
```python
st.markdown('''
<style>
.custom-column { background-color: lightblue; padding: 5px; }
</style>
''', unsafe_allow_html=True)

col1.markdown('<div class="custom-column">', unsafe_allow_html=True)
```
- Streamlit 기본 디자인을 넘어서고 싶을 때, **파이썬 코드 안에서 CSS를 문자열로 주입**
- `unsafe_allow_html=True`를 켜야 HTML로 해석됨 (기본은 텍스트로만 표시)

### ⑨ 실행 방법
- 주피터에서 `%%writefile app.py`로 코드를 파일로 저장
- 터미널에서 `streamlit run app.py`로 실제 웹앱 실행 (브라우저 자동 오픈)

## 오늘 실습 중 실제로 부딪혔던 에러와 연결

| 개념 | 오늘 겪은 에러 |
|---|---|
| `#` 주석 | 주석과 설명이 줄바꿈으로 분리되어 `NameError` 발생 |
| CSV 인코딩 | `UnicodeDecodeError` → `encoding='cp949'` 필요 |
| 파일 경로 | `souce`(오타) → `source` |
| `with` 블록 | 들여쓰기 누락으로 `IndentationError` 반복 |
| HTML 문자열 삽입 | 따옴표 위치 오류(`<'div...`), `class` 오타(`calss`) |
| 컬럼 변수 혼동 | `col1`/`col2` 복사 후 이름 수정 누락 |

**총평**: 슬라이드에 나온 개념 자체(위젯, 레이아웃, 시각화, HTML 삽입)는 오늘 실습에서 이미 실전 코드로 대부분 다뤄봤고, 막혔던 지점은 전부 개념 문제가 아니라 **오타·들여쓰기·인코딩** 같은 세부 실수였습니다.
